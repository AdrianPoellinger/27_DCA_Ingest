```python

import os
import binascii
import re
import chardet

path = "/home/renku/work/dcaonnextcloud-500gb/DigitalMaterialCopies/WeingutGantenbein/gramazio-kohler-archiv-server/036_WeingutGantenbein/04_Entwurf/060625_FinalDesign/01_Scripts/generateWalls.mel"

# --- 1) Dateiinfo (ersetzt: file, stat) ---
print("=== FILE INFO ===")
stat = os.stat(path)
print(f"Size: {stat.st_size} bytes")
print(f"Modified: {stat.st_mtime}")
print(f"Permissions: {oct(stat.st_mode)}")


# --- 2) Encoding-Erkennung ---
print("\n=== ENCODING GUESS ===")
with open(path, "rb") as f:
    data = f.read()
enc = chardet.detect(data)
print(enc)


# --- 3) Hexdump der ersten 256 Bytes (ersetzt: xxd, hexdump) ---
print("\n=== HEXDUMP (first 256 bytes) ===")
hexblock = binascii.hexlify(data[:256], sep=b" ").decode()
print(hexblock)


# --- 4) Strings-Suche (ersetzt: strings) ---
print("\n=== STRINGS FOUND (ASCII, min length 4) ===")
strings = re.findall(rb"[ -~]{4,}", data)
for s in strings[:50]:   # nur erste 50 anzeigen
    print(s.decode(errors="ignore"))


# --- 5) Anzeigen der ersten 40 Textzeilen ---
print("\n=== FIRST 40 LINES ===")
try:
    text = data.decode(enc["encoding"], errors="ignore")
    for i, line in enumerate(text.splitlines()[:40], 1):
        print(f"{i:02d}: {line}")
except Exception as e:
    print("Could not decode as text:", e)
```
